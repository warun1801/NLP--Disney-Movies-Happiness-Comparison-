```python
import pandas as pd
```

```python
movies = ['Aladdin','Cars','Finding Nemo','Frozen','Tangled','The Little Mermaid','Toy Story 3','Toy Story','Up','Zootopia']
data = {}
for c in movies:
    with open("Synopses/" + c + ".txt", "rb") as file:
        data[c] = file.read().decode("utf-8")
```

```python
pd.set_option('max_colwidth',150)

data_df = pd.DataFrame([data]).transpose()
data_df.columns = ['synopsis']
data_df = data_df.sort_index()
data_df
```

```python
data_df.synopsis.loc['Aladdin']
```

```python
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub('\(.*?\)','',text)
    text = re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\r', '', text)
    
    return text
clean = lambda x: clean_text(x)
```

```python
data_clean = pd.DataFrame(data_df.synopsis.apply(clean))
data_clean
```

```python
data_clean.synopsis.loc['Aladdin']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(data_clean.synopsis)
data_dtm = pd.DataFrame(data_cv.toarray(),columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```

```python
import pickle
```

```python
data_dtm.to_pickle("dtm.pkl")
```

```python
data_clean.to_pickle('data_clean.pkl')
pickle.dump(cv,open("cv.pkl","wb"))
```
